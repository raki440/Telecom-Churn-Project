## In churn prediction, we assume that there are three phases of customer lifecycle :

The ‘good’ phase [Month 6 & 7]
The ‘action’ phase [Month 8]
The ‘churn’ phase [Month 9]
In this case, since you are working over a four-month window, the first two months are the ‘good’ phase, the third month is the ‘action’ phase, while the fourth month is the ‘churn’ phase.

We can create new feature as total_rech_amt_data_x using total_rech_data_X and av_rech_amt_data_X to capture amount utilized by customer for data.

**Analysis:** Also as the minimum value is 1 for above columns so we can impute the NA values by 0 which is considered as no recharge 

```python
churn['total_rech_data_6'] = churn['total_rech_data_6'].replace(np.NaN,0.0)
churn['total_rech_data_7'] = churn['total_rech_data_7'].replace(np.NaN,0.0)
churn['total_rech_data_8'] = churn['total_rech_data_8'].replace(np.NaN,0.0)
```

```python
churn['av_rech_amt_data_6'] = churn['av_rech_amt_data_6'].replace(np.NaN,0.0)
churn['av_rech_amt_data_7'] = churn['av_rech_amt_data_7'].replace(np.NaN,0.0)
churn['av_rech_amt_data_8'] = churn['av_rech_amt_data_8'].replace(np.NaN,0.0)
```

```python
churn.loc[:,churn_rchg_col].describe()
```

### Filter High Value Customers


```python
##let us derive new column with av_rech-amt * total_rech_data
churn['total_rech_data_amt_6'] = churn['av_rech_amt_data_6'] * churn['total_rech_data_6']
churn['total_rech_data_amt_7'] = churn['av_rech_amt_data_7'] * churn['total_rech_data_7']
churn['total_rech_data_amt_8'] = churn['av_rech_amt_data_8'] * churn['total_rech_data_8']
churn['total_rech_data_amt_9'] = churn['av_rech_amt_data_9'] * churn['total_rech_data_9']
```

```python
churn_av_rech_6n7 = (churn['total_rech_amt_6'].fillna(0) 
+ churn['total_rech_amt_7'].fillna(0) 
+ churn['total_rech_data_amt_6'].fillna(0) 
+ churn['total_rech_data_amt_7'].fillna(0))/2
```

```python
# now we dont need columns av_rech_amt_data_x,total_rech_data_x (x = 6/7/8) , lets drop them
churn.drop(['total_rech_data_6','total_rech_data_7','total_rech_data_8','total_rech_data_9',
'av_rech_amt_data_6','av_rech_amt_data_7','av_rech_amt_data_8','av_rech_amt_data_9'],axis = 1,inplace = True)
```

```python
### take 70 percentile of the calculated average amount
percentile_70 = np.percentile(churn_av_rech_6n7, 70.0)
#percentile_70_6n7 = np.percentile(telecom_av_rech_6n7, 70.0)
print("70 percentile is : ", percentile_70)

churn_cust = churn[churn_av_rech_6n7 >= percentile_70]
#print(churn_cust)
print("Dimensions of the filtered dataset:",churn_cust.shape)
```

### Then proceed further to create churn variable after finding the high value customers

```python
# lets introduce a new column "churn", values would be either 1 (churn) or 0 (non-churn)
# we will calculate churn/non-churn based on the usage as mentioned in the problem statement
churn_cust['churn'] = np.where(churn_cust[['total_ic_mou_9','total_og_mou_9','vol_2g_mb_9','vol_3g_mb_9']].sum(axis=1) == 0, 1,0)
churn_cust.head()
```

```python
# lets find out churn/non churn percentage
churn_cust['churn'].value_counts()
churn_cust['churn'].value_counts() /len(churn_cust )*100
#observation : 92% of the customers do not churn, this might be a case of class imbalance, we will treat it later
```

```python
# lets check the null values present in the dataset
(churn_cust.isnull().sum()*100/len(churn_cust)).sort_values(ascending = False)
```

```python
### lets drop Columns with > 30% of missing values except 9th Month's columns
cols = churn_cust.columns
churn_null_perc = churn_cust.isnull().sum() * 100 / len(churn_cust)
churn_null_df = pd.DataFrame({'col_name': cols,
                                 'perc_null': churn_null_perc})

###drop the columns only it has string _9 & with missing value percentage is >=30%
drop_cols = churn_null_df.loc[(churn_null_df["col_name"].str.contains('_9')==False) & (churn_null_df["perc_null"] > 30.0)]["col_name"]
print("list of columns dropped:",drop_cols)

# lets drop these columns
churn_cust.drop(drop_cols, axis=1,inplace = True)
print(churn_cust.shape)
```

```python
## we will set intial churn=0
churn_cust['churn'] = 0
```

```python
#is_churned boolean flag for customers where above values are zero
is_churn = (churn_cust.total_ic_mou_9 == 0) & \
             (churn_cust.total_og_mou_9 == 0) & \
             (churn_cust.vol_2g_mb_9 ==0) & \
             (churn_cust.vol_3g_mb_9 ==0)
```

```python
## set all features are having is_churn True condition as 1
churn_cust.loc[is_churn, 'churn']=1
```

```python
# let us check what's the % of churned customers
100*churn_cust.churn.sum()/len(churn_cust)
```

### After tagging churners, remove all the attributes corresponding to the churn phase (all attributes having ‘ _9’, etc. in their names).



```python
churn_mon_col =  churn_cust.columns[churn_cust.columns.str.contains('_9')]
print(len(churn_mon_col))
print(churn_cust.shape)
```

```python
###let us remove all the columns or attributes corresponding to the churn phase which are having _9 in their names.
churn_cust.drop(churn_mon_col, axis=1, inplace=True)
```
